```python
zips=big_data['Incident Zip']
is_close=zips.str.startswith('0')|zips.str.startswith('1')
zips[~(is_close)&zips.notnull()]
```

```python
def aoe():
    li=[]
    for i in zips:
        if type(i)==str:
            li.append(i)
        elif type(i)==float:
            li.append(str(i))
    return li
```

```python
big_data['Incident Zip']=aoe()
```
