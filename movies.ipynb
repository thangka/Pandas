# MERGE &JOIN

```python
df1 = pd.DataFrame({'Name':['Ar','Ka','Ha','Rak','Na'],
                   'Age':[19,20,20,21,19],
                   'Department':['CS','EC','CS','IT','IT']})

df2=pd.DataFrame({'Name':['Ar','Ha','Rak','Ram','Ka'],
                 'Courses':['Machine','OOPS','Web','OOPS','Data']})

display(df1, df2)
```

```python
pd.merge(df1,df2,on='Name',how='inner')
```

```python
set(df1['Name'])&(set(df2['Name']))
```

```python
pd.merge(df1,df2,on='Name',how='left')
```

```python
pd.merge(df1,df2,on='Name',how='right')
```

```python
pd.merge(df1,df2,on='Name',how='outer')
```

```python
left = df1.set_index("Name")

right = df2.set_index("Name")

result = left.join(right, lsuffix="_l", rsuffix="_r")
result
```

# eval ### tao ra cot moi

```python
result.eval('a_age=Age*4/3')
```

# query ### thay the mask()

```python
result.query('Age>=20 | a_age>25')
```
